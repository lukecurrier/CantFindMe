```python
import seaborn
import matplotlib.pyplot as plt
import numpy as np 
```

```python
mistral = 0.0
qwen_01 = 0.04
qwen_02 = 0.167
qwen_03 = 0.4444
cantfindme_ensemble = 0.0
cantfindme_compression = 0.425
cantfindme_perturb = 0.3
hidenseek = 0.72
```

```python
labels = ['mistral', 'qwen01', 'qwen02', 'qwen03', 'ensembling', 'compression', 'perturbation', 'hidenseek']
values = [mistral, qwen_01, qwen_02, qwen_03, cantfindme_ensemble, cantfindme_compression, cantfindme_perturb, hidenseek]
colors = ['lightblue', 'lightblue', 'lightblue', 'lightblue','lightgreen', 'pink', 'lightgreen', 'lightcoral']

plt.figure(figsize=(12, 6))  
plt.bar(labels, values, width=0.4, color=colors)
for i, value in enumerate(values):
    plt.text(i, value + 0.02, f'{value*100:.1f}%', ha='center')

plt.ylim(0, 0.8)
plt.title('Evolutionary Fingerprinting Success Rates')
plt.show()
```
